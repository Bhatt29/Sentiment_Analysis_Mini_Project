```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv('/content/twitter.csv')
df
```

```python
df = df.dropna(subset=['id'])
df = df.dropna(subset=['tweet'])
```

```python
df = df.drop(columns=['id'])
```

```python
df
```

```python
import re
df['tweet'] = df['tweet'].apply(lambda x: re.sub(r'Ã°ÂŸÂ“Â±!!!¦  ', ' ', x))
df['tweet'] = df['tweet'].apply(lambda x: re.sub(r'[@#]\w+', ' ', x))
df['tweet'] = df['tweet'].apply(lambda x: re.sub(r'\s+', ' ', x))
```

```python
df
```

```python
from sklearn.feature_extraction.text import CountVectorizer

# Create a CountVectorizer object
vectorizer1 = CountVectorizer(binary=True,stop_words='english')
vectorizer2=CountVectorizer(binary=False,stop_words='english')
# Fit and transform the text data
X1 = vectorizer1.fit_transform(df['tweet'])
X2=  vectorizer2.fit_transform(df['tweet'])
y=df['label']
```

```python
import nltk
from nltk.corpus import stopwords
nltk.download('stopwords')
print(stopwords.words('english'))
```

```python
from sklearn.model_selection import train_test_split
X_train1, X_test1, y_train1, y_test1 = train_test_split(X1, y, test_size=0.2, random_state=42)
X_train2, X_test2, y_train2, y_test2 = train_test_split(X2, y, test_size=0.2, random_state=42)
```

```python
from sklearn.naive_bayes import BernoulliNB, MultinomialNB
model1=BernoulliNB()
```

```python
model1.fit(X_train1,y_train1)
```

```python
model2=MultinomialNB()
```

```python
model2.fit(X_train1,y_train1)
```

```python
prediction1=model1.predict(X_test1)
prediction2=model2.predict(X_test2)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test1,prediction1)
```

```python
accuracy_score(y_test2,prediction2)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.pipeline import make_pipeline
model=make_pipeline(TfidfVectorizer(), MultinomialNB() )
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(df['tweet'],df['label'],test_size=0.25,random_state=12)
```

```python
model.fit(xtrain,ytrain)
```

```python
prediction3=model.predict(xtest)
```

```python
accuracy_score(ytest,prediction3)
```

Conclusion: For the twitter data TFIDFVectorizer is used compared to MultinomialNB and BinomialNB as TFIDF gives more accuracy than other two.

```python
# Function to predict sentiment of a user-provided tweet
def predict_tweet_sentiment(tweet):

    # Predict sentiment using the trained model
    prediction = model.predict([tweet])
    return prediction[0]

# Get user input
user_tweet = input("Enter a tweet: ")
sentiment = predict_tweet_sentiment(user_tweet)
print("Predicted Sentiment:", sentiment)
```
